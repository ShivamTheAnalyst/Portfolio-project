```python
import pandas as pd
import seaborn as sns
import numpy as np

from matplotlib import pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matlpotlib inline
 
```

```python
# READING THE DATA

df = pd.read_csv('movies.csv')
```

```python
# LOOKING AT THE DATA
df
```

```python
# Let's see if there is any missing data
for col in df.columns:
    missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,round(missing*100)))
```

```python
# looking at data types for our columns 
df.dtypes
```

```python
# Cleaning the data.
```

```python
#Changing bugdet and gross column from float to int64 for better appearance.
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
# The column types of gross and budget are changed.
```

```python
# Right now the released is object data type but we will change it to string so that we can full specific data from it. 
# Making a new column 'Correct_year' whose values are taken from released column first four strings.
df['Correct_year'] =  df['released'].astype(str).str[:4]
```

```python
df.head()
```

```python
# Sorting by 'gross' column.

df.sort_values(by=['gross'], inplace = False, ascending = False )
```

```python
# Plot the budget vs gross using SEABORN. The line will show the correlation between budget and gross earnings,here it is high.

sns.regplot(x='budget',y='gross',data= df, scatter_kws={'Color':'red'}, line_kws={'Color':'black'})
```

```python
# Let's start looking at correlation 
df.corr(method='pearson') #-> This will only work on numeric fields (Numeric data types) ->
# method used here By Default -pearson correlation, others are kendall, spearman.
```

```python
# it's really hard to look in the above numbers(The pearson correlation matrix), so we will visualize this .
#assigning it to a variable

Correlation_matrix = df.corr(method = 'pearson')
sns.heatmap(Correlation_matrix, annot = True )  
plt.title("Correlation Matrix from numeric features")
plt.show() # idk y im not able to put labels
```

```python
# Looking at company column for correlation.
# As we can see that company column type is not numeric but we can convert , so we are doing the same

df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes # Idk why we have done this-do reserach here.
        
        

```

```python
df_numerized
```

```python
# looking at correlation again.
new_correlation_matrix = df_numerized.corr(method='pearson')
```

```python
new_correlation_matrix
```

```python
#Adjusting the size of the plot for clean representation of the heatmap.
from matplotlib import rcParams

# figure size in inches
rcParams['figure.figsize'] = 11.7,8.27
```
